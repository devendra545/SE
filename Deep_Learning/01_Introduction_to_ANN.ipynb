# Machine Learning Landscape

## Machine Learning

The field of study that gives computers the ability to learn without being explicitly programmed.

## Common Machine Learning Applications

- Housing Price Prediction
- Autonomous Vehicles
- Message sentiment Analysis
- Spam email filtering
- Recommendation Engines
- Image Recognition

## Machine Learning: Key Concepts

### The Data & The Process

**Focus: Structured Data**

- Data in Tables(rows & columns)
- Predict a target from features

**Method: Manual Feature Engineering**

- Human Experts design features
- Model learns these engineered features
- Success depends entirely on feature quality

### The Models

**Linear Models**

- *Core Idea:* Assumes linear relationships
- *Examples:* Linear/Logistic regression
- *Traits:* Fast, interpretable but simple

**Tree Based Models**

- *Core Idea:* Learns 'if-then-else' rules
- *Examples:* Gradient boosting / Random forest
- *Traits:* Captures non-linearity powerful

## Limitations of Traditional Machine Learning



- **The Feature Engineering Bottleneck**: Time-consuming, requires domain expertise, and
is often suboptimal.
- **Inability to handle high-dimensional data**: Struggles with images, audio, or raw text.
- **Limited representation learning**: Learns shallow patterns, not deep, hierarchical
features.
- **Performance plateaus**: Adding more data yields diminishing returns after a certain point.

# The Deep Learning Revolution

## The Shift to Deep Learning



### Deep Learning: The Core Idea

- Models learn directly from raw data.
- Handles images, text, and audio.
- No manual feature engineering.

### Why this paradigm shifts?

- Solves the feature engineering bottleneck.
- Automates representation learning.
- Unlocks new, complex capabilities.

## Deep Learning Applications

- Computer Vision
- Natural Language Processing
- Speech Recognition and Audio Synthesis
- Game Playing & Robotics

## Key Differences: ML vs DL

|**Parameter**| **Machine Learning**|**Deep Learning**|
|---|---|---|
|**Feature Extraction:**| Requires manual engineering| Learns features automatically|
|**Representation:**|Learns shallow patterns|Builds Hierarchical features|
|**Data Scalability:**| Scale well on small to medium dataset|Excels with very large datasets|
|**Training Paragigm**|-|Enables end-to-end learning|

## What enabled the DL Revolution?

A convergence of three key factors created the perfect storm.

|**Big Data**|**Computational Power**|**Better Algorithms**|
|---|---|---|
|Massive Labeled datasets|Rise of GPUs (parallel computing)|Better Algorithms|
|The "fuel for hungry models.|Made deep learning feasible|e.g., CNNs, Transformers|
|e.g., ImageNet, Wikipedia|Drastically cut training time|Smarter training methods|
