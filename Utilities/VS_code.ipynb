# Hiding Files from Explorer

- Open command Palette -> Preferences: Open settings(JSON) -> add

```json
    "files.exclude": {
    "**/.venv": true,
    "**/__pycache__": true,
    "**/.git*": true
```

# Shortcuts

## Configuring Shortcuts

- Open Command Palette
- Preferences: Open keyboard shortcuts (JSON)

```json
{
  "key": "ctrl+shift+z",
  "command": "workbench.action.toggleZenMode"
}
```

## Useful Shortcuts

- ***Ctrl + B***: Toggle the sidebar
- ***Ctrl + P***: Open any file
- ***Ctrl + `***: Open Terminal
- ***Ctrl + A***: Select All
- ***Ctrl + D***: Select next occurrence
- ***Alt + Click***: Add multiple cursors
- ***Ctrl + F***: Find in File
- ***Ctrl + H***: Replace in File
- ***Ctrl + Shift + F***: Find in All File
- ***Ctrl + Tab***: Switch between recent tab
- ***Ctrl + Shift + Tab***: Reverse switch
- ***Ctrl + G***: Go to line number
- ***Ctrl + Shift + E***: Toggle File Explorer
- ***Ctrl + Shift + o***: Go to Specific symbol
- ***Ctrl + Shift + Z***: Toggle Zen mode (custom)
- ***Ctrl + Shift + P***: Open Command Palette
- ***Ctrl + Shift + V***: Open Preview
- ***Ctrl + Shift + X***: Open Extensions
- ***Ctrl + Shift + K***: Delete the entire line (on which cursor is placed)
- ***Ctrl + W***: Close any tab

# Snippets

- Step-1: Ctrl + Shift + P
- Step-2: Snippets
- Step-3: Select snippet type or file (generally default)
- Step-4: Give snippet file name: snippet.code-snippet

```json
{
    "theta": {
        "prefix": "/theta",
        "body": "$\\theta$"
    }
}
```

- prefix: what we type
- body: what gets inserted
